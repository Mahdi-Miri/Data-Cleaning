```python
import nltk
import pandas as pd
import re

nltk.download('punkt')
```

```python
splits = {'train': 'plain_text/train-00000-of-00001.parquet', 'test': 'plain_text/test-00000-of-00001.parquet', 'unsupervised': 'plain_text/unsupervised-00000-of-00001.parquet'}
df = pd.read_parquet("hf://datasets/Kwaai/IMDB_Sentiment/" + splits["train"])
```

```python
df.head()
```

```python
df.shape
```

```python
df_text = df['text']
```

```python
df_text.head()
```

```python
df_text.shape
```

```python
data = df_text.to_string()
```

```python
data
```

# Remove Newlines And Tabs

```python
clean_data = data.replace('\n',' ')
clean_data= clean_data.replace('\t',' ')
clean_data = re.sub(r"\s+", "", clean_data)
clean_data
```

# Remove Punctuation - Unicode Characters - Special Characters

```python
clean_data = clean_data.encode("ascii","ignore")
clean_data = clean_data.decode()
clean_data
```
